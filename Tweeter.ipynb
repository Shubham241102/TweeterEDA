```python
df.groupby(['Tweet_ID', 'Retweets'   , 'Timestamp' ]).size().reset_index()
df.head()
```

```python
Retweets =df.groupby([ 'Tweet_ID' , 'Retweets']).size().reset_index().rename(columns={0:'rate_count'})
```

```python
Retweets
```

# Observation

From Above Operation we can observe that 
1. Tweet_ID 01 received 2 Retweets.
2. Tweet_ID 9996 received 81 Retweets.
3. Tweet_ID 10000 received 65 Retweets.

```python
Retweets.head()
```

```python
Tweet_ID =df.groupby([ 'Retweets' , 'Likes']).size().reset_index().rename(columns={0:'rate_count'})
```

```python
plt.rcParams['figure.figsize']=(30,10)
sns.barplot(x='Retweets', y= 'Likes', data=Tweet_ID  )
```
